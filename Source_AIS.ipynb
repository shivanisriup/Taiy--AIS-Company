```python
import requests
from bs4 import BeautifulSoup
import re
import dateutil
```

```python
result = requests.get("https://www.aishub.net/stations")
```

```python
result
```

```python
src = result.text
document = BeautifulSoup(src, 'lxml')
```

```python
table = document.find("table")
```

```python
table
```

```python
rows = table.find_all("tr")  # Note: this works because find_all is resursive by default
```

```python
rows
```

```python
#For getting all information from 1 page 
for row in rows[2:-1]:
    cells = row.find_all(["td", "th"])
    cells_text = [cell.get_text(strip=True) for cell in cells]
    (ID, Status, Uptime, Country, Location, Ships, Distinct, Contributor) = cells_text    
    print(f"{ID}, {Status}, {Uptime}, {Country}, {Location}, {Ships} , {Distinct} , {Contributor}")
```

```python
#For fetching all information from page 1 to 49. 
#Similarily if we want to scrap all 936 pages then just change number 50 to 937.You will able to scrap all information
for page in list(range(1,50)):
  result = requests.get("https://www.aishub.net/stations?page={}".format(page))
  src = result.text
  document = BeautifulSoup(src, 'lxml')
  table = document.find("table")
  rows = table.find_all("tr")  # Note: this works because find_all is resursive by default
  for row in rows[2:-1]:
    cells = row.find_all(["td", "th"])
    cells_text = [cell.get_text(strip=True) for cell in cells]
    (ID, Status, Uptime, Country, Location, Ships, Distinct, Contributor) = cells_text    
    print(f"{ID}, {Status}, {Uptime}, {Country}, {Location}, {Ships} , {Distinct} , {Contributor}")
```
